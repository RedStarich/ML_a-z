```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X, y)
y_pred = regressor.predict(X)
```

```python
regressor.predict([[6.5]])
```

bad prediction, actual ~ 160000
SVR showed more accurate result, thus, SVR is better for such regression datasets
Decision tree is better if there are a lot of features (here we have only 1 feature), so this model is better to be applied to the "Titanic" and "House SalePrice" tasks

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Salary expectation predictor')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

The graph is as smooth as we want it to be
According to the graph, Decision tree used dataset results and split it according to a specific range
The graph is shaped as a ladder-like regression

```python
from sklearn.metrics import r2_score
r2_score(y, y_pred)
```

R2 score is 100%, which means the model works perfect on the trained data, but this score is not applicable for new data
