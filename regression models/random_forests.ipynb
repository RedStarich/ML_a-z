```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators= 10, random_state=0)
regressor.fit(X, y)
y_pred = regressor.predict(X)
```

```python
regressor.predict([[6.5]])
```

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Salary expectation predictor')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

The result of Random Forest is very close to the image of Decision tree
The graph resembles the same ladder-like characteristic
However, random forest seems to get the lower-bound value of each 'step' of the 'ladder'

```python
from sklearn.metrics import r2_score
r2_score(y, y_pred)
```

R2 score showed the accuracy of 97%, which is amazing result for the model
